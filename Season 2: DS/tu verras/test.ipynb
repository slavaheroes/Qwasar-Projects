```python
from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
```

```python
#load data
dataset = load_boston()
```

```python
#converting data in pandas dataframe
def load_dataframe(dataset):
    df = pd.DataFrame(data= np.c_[dataset['data']], columns= dataset['feature_names'])
    #added extra column
    df["MEDV(target)"] = dataset['target']
    
    return df

boston_dataframe = load_dataframe(dataset)

boston_dataframe[:20]
```

```python
#count null values using isnull() and sum() functions
def count_missing_values(boston_dataframe):
    return boston_dataframe.isnull().sum()

print(count_missing_values(boston_dataframe))
#here is no null values
```

```python
import matplotlib.pyplot as plt

def plot_histograms(df):
    fig, axs = plt.subplots(4,4)
    fig.tight_layout()
    col_names = df.columns
    k = 0
    for i in range(4):
        for j in range(4):
            if k < len(col_names):
                axs[i,j].hist(df[col_names[k]])
                axs[i,j].set_title([col_names[k]])
            k += 1
    
    plt.show()


plot_histograms(boston_dataframe)
```
